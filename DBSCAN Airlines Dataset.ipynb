Rule for missing value imputation:

If NA values are between 1 to 5%, drop na (rows)

If NA value are 5 to 40%, fillna(replace)

If NA values are greater than 50% in a column, drop that entire column

# Model Building

Feature scaling

```python
sc = StandardScaler()

x = sc.fit_transform(df)
x
```

```python
db = DBSCAN(eps=0.5, min_samples=3)  ##cosidering eps = 0.5 and min_sample =3 then we will do hyperparameter tunning

y = db.fit_predict(x)
y
```

```python
plt.scatter(x[:,0],x[:,4],c = y,cmap = 'rainbow')
```

```python
silhouette_score(x,y)
```

# Hyperparameter tunning

Selecting Min_samples:¶

1.Number of columns + 1

Number of columns * 2

Always keep min_sample at least 3

```python
neigh = NearestNeighbors(n_neighbors=3).fit(x)
```

```python
d,index = neigh.kneighbors(x)
```

```python
d
```

```python
distance = np.sort(d[:,1])
distance
```

```python
plt.plot(distance)
plt.axhline(y=1, linestyle = '--', color = 'red')
plt.xlabel('Distance')
plt.ylabel('EPS')
plt.title('K-Distance Graph')
```

# Evaluate model 

here we are changing the eps value to increase models performance i,e, schellotte value near to 1

```python
db = DBSCAN(eps=2,min_samples=3)
yp = db.fit_predict(x)
yp
```

```python
silhouette_score(x,yp)
```
